```python
df = pd.read_excel('data/pa.xlsx', dtype = {'cum':int})
df.head()
```

```python
df['DateTime'] = df['mmdd'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
```

```python
df.info()
df.head()
```

```python
def create_geojson_features(df):
    print('> Creating GeoJSON features...')
    features = []
    for _, row in df.iterrows():
        feature = {
            'type': 'Feature',
            'geometry': {
                'type':'Point', 
                'coordinates':[row['lng'],row['lat']]
            },
            'properties': {
                'time': row['DateTime'].date().__str__(),
                'style': {'color' : row['color']},
                'icon': 'circle',
                'iconstyle':{
                    'fillColor': row['color'],
                    'fillOpacity': 0.8,
                    'stroke': 'true',
                    'radius': 7
                }
            }
        }
        features.append(feature)
    return features
```

```python
def make_map(features):
    print('> Making map...')
    coords_belgium=[40.4406, -79.9959]
    pollution_map = folium.Map(location=coords_belgium, control_scale=True, zoom_start=8)

    TimestampedGeoJson(
        {'type': 'FeatureCollection',
        'features': features}
        , period='P1D'
        , add_last_point=True
        , auto_play=False
        , loop=False
        , max_speed=1
        , loop_button=True
        , date_options='YYYY/MM/DD'
        , time_slider_drag_update=True
    ).add_to(pollution_map)
    print('> Done.')
    return pollution_map
```

```python

def plot_pollutant(pollutant_ID):
    

    features = create_geojson_features(df)
    return make_map(features), df
```
