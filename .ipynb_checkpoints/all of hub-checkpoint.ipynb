```python
color_scale = np.array(['#1a9850','#fee08b','#f46d43','#a50026','#1a1a1a']) # 'bin_edges': np.array갯수 +1 로 np.array세팅해야함.
sns.palplot(sns.color_palette(color_scale))
```

```python
meta = pd.read_csv('data/meta_data.csv', dtype={'fac_id':'object'})
meta.head()
meta.info()
```

```python

def load_data(hub_ID):
    print('> Loading data...')
    date_vars = ['dsps_dtm', 'change_dtm'] #parse_date에 사용, date_parser를 날짜 형태로 받을 것임.
    filename = 'data/raw/dsps_' + str(hub_ID) + '_0125-0129.csv'
    agg_ts = pd.read_csv(filename, parse_dates=date_vars, date_parser=pd.to_datetime,dtype={'fac_id':'object'})
    return agg_ts
```

```python
def color_coding(poll, bin_edges):    
    idx = np.digitize(poll, bin_edges, right=True) #연속형 변수 이산형화(discretization, poll의 값으로 bin_edges), 
    # np.digitize(data, bins)
    return color_scale[idx]

def prepare_data(df, hub_ID):
    print('> Preparing data...')
    df = df.reset_index().merge(meta, how='inner', on='fac_id').set_index('change_dtm') # merging
    df['color'] = df.cum.apply(color_coding, bin_edges=hubs[hub_ID]['bin_edges'])# 여기서 bin_edges를 입력 받아서 위에서 씀.
    # cum값으로 
    
    # lambda 사용 데이타 타입 변경 apply(lambds)
    df['cum_int']= df['cum'].apply(lambda x: int(x) if isinstance(x, float) else x) # float면 int로, 아니면 그대로
    
    cols = ['fac_name','cum_int'] # 합칠 컬럼 변수로 지정
    df['name-cum_int'] = df[cols].apply(lambda x: '\n' + ': '.join(x.dropna().astype(str)), axis=1) #컬럼 합침 name: cum

    return df

```

```python
def create_geojson_features(df):
    print('> Creating GeoJSON features...')
    features = []
    for _, row in df.iterrows(): # df 의 레이블(여기서는 change_dtm) 과 row(각 값)
        feature = {
            'type': 'Feature',
            'geometry': {
                'type':'Point', 
                'coordinates':[row['Longitude'],row['Latitude']] # 순서 주의
            },
            'properties': {
                'time': row['dsps_dtm'].date().__str__(),
                'style': {'color' : row['color']},
                'icon': 'circle',
                'popup': row['name-cum_int'], # 팝업에 나올것
                'iconstyle':{
                    'fillColor': row['color'],
                    'fillOpacity': 0.8,
                    'stroke': 'true',
                    'radius': 7
                }
            }
        }
        features.append(feature)
    return features
```

```python
def make_map(features, hub_ID):
    print('> Making map...')
    coords_point = hubs[hub_ID]['coords_point'] # hubs딕셔너리에서 찾아오기
    hub_map = folium.Map(location=coords_point, control_scale=True, zoom_start=6)
    folium.Marker(hubs[hub_ID]['hub_location'], popup='HUB',icon=folium.Icon(color='blue', icon='home')).add_to(hub_map)
    # 허브 위치 추가

    TimestampedGeoJson(
        {'type': 'FeatureCollection',
        'features': features}
        , period='P1D' #day별로
        , add_last_point=True
        , auto_play=False
        , loop=False
        , max_speed=0.6
        , loop_button=True
        , date_options='YYYY/MM/DD'
        , time_slider_drag_update=True
    ).add_to(hub_map)
    print('> Done.')
    return hub_map
```

```python
def plot_hub(hub_ID):
    print('Mapping {} dsps in 0128-0130'.format(hubs[hub_ID]['name'])) # {} 에 format 정보 입력
    df = load_data(hub_ID)
    df = prepare_data(df, hub_ID)
    features = create_geojson_features(df)
    return make_map(features, hub_ID), df
```

```python
# 하기와 같이 해도 됨
def plot_hub_(hub_ID):
    print('Mapping %s dsps in 0128-0130' % hubs[hub_ID]['name'])
    
plot_hub_(1)
```
